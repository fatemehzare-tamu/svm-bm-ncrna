```python
import pickle

# Load X arrays
with open("X_train.pkl", "rb") as f:
    X_train = pickle.load(f)

with open("X_test.pkl", "rb") as f:
    X_test = pickle.load(f)

# Load y arrays
with open("y_train.pkl", "rb") as f:
    y_train = pickle.load(f)

with open("y_test.pkl", "rb") as f:
    y_test = pickle.load(f)


print("Original X_train shape:", X_train.shape)
print("Original y_train shape:", y_train.shape)
print("Original X_test shape:", X_test.shape)
print("Original y_test shape:", y_test.shape)
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import linear_kernel, rbf_kernel

# Combine datasets
X_all = np.vstack([X_train, X_test])
y_all = np.concatenate([y_train, y_test])
X1 = X_all[:, 0].reshape(-1, 1)

# Helper: plot class distribution
# -----------------------------
def plot_distribution(values, labels, title):
    plt.figure(figsize=(7,4))
    plt.hist(values[labels == 1], bins=40, alpha=0.6, label='Class +1')
    plt.hist(values[labels == -1], bins=40, alpha=0.6, label='Class -1')
    plt.title(title)
    plt.xlabel("Value")
    plt.ylabel("Count")
    plt.legend()
    plt.grid(True)
    plt.show()

# -----------------------------
# 1. Original distribution
# -----------------------------
plot_distribution(X1.flatten(), y_all, "Original Feature 1 Distribution")

# -----------------------------
# 2. Linear kernel transformation
# For 1D: linear kernel k(x,y)=x*y  identity mapping is enough
# -----------------------------
X1_linear = X1.flatten()   # no change needed
plot_distribution(X1_linear, y_all, "After Linear Kernel (identity)")

# -----------------------------
# 3. RBF kernel transformation (1D)
#
# k(x,z) = exp(- (x - z)^2)
# We choose z = mean(X1) as a single landmark
# This produces a 1D nonlinear transform without huge memory cost
# -----------------------------
gamma = 0.5
z = np.mean(X1)

X1_rbf = np.exp(-gamma * (X1.flatten() - z)**2)

plot_distribution(X1_rbf, y_all, f"After RBF Kernel (landmark = mean, ={gamma})")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------
# Step 1: Combine datasets
# ---------------------------
X_all = np.vstack([X_train, X_test])
y_all = np.concatenate([y_train, y_test])

# ---------------------------
# Step 2: Balance classes
# ---------------------------
np.random.seed(42)

# Find indices for each class
idx_pos = np.where(y_all == 1)[0]
idx_neg = np.where(y_all == -1)[0]

# Choose same number of samples for each class
n_samples_per_class = min(len(idx_pos), len(idx_neg))
sel_pos = np.random.choice(idx_pos, n_samples_per_class, replace=False)
sel_neg = np.random.choice(idx_neg, n_samples_per_class, replace=False)

# Combine selected indices and shuffle
selected_idx = np.concatenate([sel_pos, sel_neg])
np.random.shuffle(selected_idx)

X_bal = X_all[selected_idx]
y_bal = y_all[selected_idx]

# ---------------------------
# Step 3: Add Gaussian noise
# ---------------------------
noise_level = 0
np.random.seed(32)
X_bal_noisy = X_bal + noise_level * np.random.randn(*X_bal.shape)

# ---------------------------
# Step 4: Shuffle and split back to train/test
# ---------------------------
np.random.seed(42)
perm = np.random.permutation(len(y_bal))
X_bal_noisy = X_bal_noisy[perm]
y_bal = y_bal[perm]

# For example, 70% train, 30% test
split = int(0.7 * len(y_bal))
X_train_new = X_bal_noisy[:split]
y_train_new = y_bal[:split]
X_test_new = X_bal_noisy[split:]
y_test_new = y_bal[split:]

print("New X_train shape:", X_train_new.shape)
print("New y_train shape:", y_train_new.shape)
print("New X_test shape:", X_test_new.shape)
print("New y_test shape:", y_test_new.shape)

# ---------------------------
# Step 5: Plot distributions per class
# ---------------------------
plt.figure(figsize=(12,5))

for i, (X_set, y_set, title) in enumerate(zip(
    [X_train_new, X_test_new], 
    [y_train_new, y_test_new],
    ['Train', 'Test']
)):
    plt.subplot(1,2,i+1)
    for cls, color in zip([-1,1], ['red','blue']):
        plt.hist(X_set[y_set==cls][:,1], bins=50, alpha=0.5, color=color, label=f'Class {cls}', density=True)
    plt.title(f'{title} Distribution (Feature 1)')
    plt.xlabel('Feature value')
    plt.ylabel('Density')
    plt.legend()

plt.tight_layout()
plt.show()
```
